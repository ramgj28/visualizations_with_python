```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titanic_data=r"C:\Users\Ramgopal.J\Desktop\NLP\my projs\titanic\titanic_train.csv"
df=pd.read_csv(titanic_data,index_col="PassengerId",usecols=["PassengerId","Survived","Pclass","Age","SibSp","Parch","Fare"],nrows=50)
```

# 1.LINEPLOT

```python
sns.set_style("ticks")
plt.figure(figsize=(16,6))
plt.title("TITANIC PASSENGERS")
sns.lineplot(data=df)
```

# 2.BARPLOT

```python
plt.figure(figsize=(16,6))
plt.title("TITANIC PASSENGERS")
sns.barplot(y=df["Age"],x=df.index)
```

# 3.HEATMAP

```python
df1=pd.read_csv(titanic_data,index_col="PassengerId",usecols=["PassengerId","Age"],nrows=100)
df1
```

```python
plt.figure(figsize=(160,60))
sns.heatmap(data=df1,annot=True)
#dataset not suitable for heatmap
```

# 4.SCATTERPLOTS

```python
sns.scatterplot(y="Fare",x="Age",data=df)
```

```python
sns.regplot(y="Fare",x="Age",data=df)
```

```python
sns.scatterplot(y="Fare",x="Age",hue="Survived",data=df)
```

```python
sns.lmplot(y="Fare",x="Age",hue="Survived",data=df)
```

```python
sns.swarmplot(y="Fare",x="Survived",data=df)
```

# HISTOGRAM

```python
sns.distplot(a=df["Age"],kde=False)
```

```python
sns.kdeplot(data=df["Age"], shade=True)
```

```python
sns.jointplot(x="Age",y="Survived",data=df,kind="kde",shade=True)
```

# To set themes-
USE sns.set_style    
it has 5 themes
(1)"darkgrid", (2)"whitegrid", (3)"dark", (4)"white", and (5)"ticks"

```python
sns.set_style("darkgrid")
sns.swarmplot(y="Fare",x="Survived",data=df)
```

```python
sns.set_style("whitegrid")
sns.swarmplot(y="Fare",x="Survived",data=df)
```
